```python
# ✅ Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style='whitegrid')
```

```python
# ✅ Step 2: Load Dataset
df = pd.read_csv("/Wholesale_Customers_Data (4).csv")  # make sure this file is uploaded
df.head()
```

```python
# ✅ Step 3: Preliminary Cleaning

# 1. Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# 2. Convert 'Channel' and 'Region' to categorical (for better visualization)
df['Channel'] = df['Channel'].astype(str)
df['Region'] = df['Region'].astype(str)
```

```python
# ✅ Step 4: Simple Transformation

# Create a new feature: Total Spend across all categories
df['Total_Spend'] = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].sum(axis=1)
df[['Total_Spend']].head()
```

```python
# 📊 1. Distribution of Total Spend
plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Spend'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Spend')
plt.xlabel('Total Spend')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# 📊 1. Distribution of Total Spend
plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Spend'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Spend')
plt.xlabel('Total Spend')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# 📊 3. Average Spend per Product Category by Region
avg_spend_by_region = df.groupby('Region')[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']].mean()

plt.figure(figsize=(14, 6))
avg_spend_by_region.T.plot(kind='bar')
plt.title('Average Spend per Category by Region')
plt.ylabel('Average Spend')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```
